## Import a Sparse Image

The first step will be to import a sparse image for us to binarize.

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
soccer_ball = np.flipud(rgb2gray(mpimg.imread("images/soccer_ball.jpeg")))
soccer_ball = np.array(soccer_ball < 230, dtype='int')
imshow(soccer_ball, cmap='binary')
```

```python
soccer_ball_resample = np.array(soccer_ball[10:210, 17:217].reshape(25, 8, 25, 8).mean(3).mean(1) > 0.5, dtype = 'int')
imshow(soccer_ball_resample, cmap='binary')
```

```python
smaug = np.flipud(rgb2gray(mpimg.imread("images/smaug.jpg")))
smaug = np.array(smaug < 100, dtype='int')
imshow(smaug, cmap='binary')
```

```python
def l1min_greedy_bin(A, b, e=0):
    """
    Greedy binary l1 minimization
    """
    m = A.shape[0]
    n = A.shape[1]
    
    sol = np.zeros((n,), dtype='int')
    err = np.sum((np.dot(A, sol) - b)**2)
    while err > e :
        
        (zero_idx,) = np.nonzero(sol == 0) # index zero elements of sol
        test_errs = np.zeros_like(zero_idx)
        # Test each element of sol to see if it improves the error
        for (idx, elem) in enumerate(zero_idx):
            sol[elem] = 1
            test_errs[idx] = np.sum((np.dot(A, sol) - b)**2)
            sol[elem] = 0
        # Pick the element of sol that gives the lowest error
        best = min(zip(test_errs, zero_idx))
        print best,
        # If it has a bigger error break
        if best[0] > err:
            return sol
        else: # If not it's our new error and set that element to 1
            err = best[0]
            sol[best[1]] = 1
        
    return sol
        
        
        
```
