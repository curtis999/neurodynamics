## Import a Sparse Image

The first step will be to import a sparse image for us to binarize.

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
soccer_ball = np.flipud(rgb2gray(mpimg.imread("soccer_ball.jpeg")))
soccer_ball = np.array(soccer_ball < 230, dtype='int')
imshow(soccer_ball, cmap='binary')
```

```python
smaug = np.flipud(rgb2gray(mpimg.imread("smaug.jpg")))
smaug = np.array(smaug < 100, dtype='int')
imshow(smaug, cmap='binary')
```

```python
trbrd = np.flipud(rgb2gray(mpimg.imread("Treebeard1.jpg")))
trbrd = np.array(trbrd < 50, dtype='int')
imshow(trbrd, cmap='binary')
```

## Perform a Wavelet Decomposition

```python
import pywt
```

```python
sbvec = np.reshape(trbrd, (trbrd.size,))
```

```python
cA, cD = np.array(pywt.dwt(sbvec, 'db1'))
```

```python
fig, ax = plt.subplots()
ax.plot(cD)
```
