## Problem 1

```python
# setting values
R = 10.                          # kOhm
C = 1.                           # uF
V_th = 20.                       # mV
E_m = 5.
V_reset = 0.
```

### 1.1

Current conservation around the loop gives
$$C\frac{dV}{dt} = I_{ext} - I_L = I_{ext} - \frac{1}{R}(V - E_m)$$
Multiplying by $R$ and rearranging a bit,
$$RC\frac{dV}{dt} = - ( V - (RI_{ext} + E_m))$$
and identifying $\tau = RC$ and $V_{\infty} = RI_{ext} + E_m$ we recognize the form
$$\frac{dV}{dt} = -\frac{(V - V_{\infty})}{\tau}$$
This is a linear first-order inhomogeneous ODE.  The membrane voltage will exponentially approach the equilibrium value with time constant $\tau$ with the added wrinkle that if $V_m$ crosses $V_{thresh}$ we reset it to $V_{reset} = 0$. Along the approach to $V_{\infty}$, the voltage is described by,
$$V_m(t) = (V_m(t=0) - V_{\infty})e^{-t/\tau} + V_{\infty}$$
until we reach $V_{th}$

### 1.2

```python
dt = 0.1
t = scp.arange(0, 250+ dt, dt)

I_ext = 0
tau = R * C


def V_nospikes(V_0, I_ext, t):
    V_inf = R * I_ext + E_m
    return (V_0 - V_inf) * scp.exp(-1 * t / tau) + V_inf

def first_spike(V):
    for idx, v in enumerate(V):
        if v >= V_th:
            break
    return idx


def V_LIF(V_0, I_ext, t):
    spike_times = []
    V = V_nospikes(V_0, I_ext, t)
    while any(V >= V_th):
        spk_idx = first_spike(V)
        spike_times.append(t[spk_idx])
        V[spk_idx:] = V_nospikes(V_reset, I_ext, t[spk_idx:] - t[spk_idx])
    return V, spike_times
    
```

```python
Is = [1.5, 2.5, 3.5]
fig1, axes1 = plt.subplots(3 , 1, figsize=(6, 1.5 * len(Is)))

for idx, ax in enumerate(axes1):
    V, temp = V_LIF(V_reset, Is[idx], t)
    ax.plot(t, V)
    ax.set_title(r'$I_{ext}=$'+str(Is[idx]))
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('Voltage (mV)')

fig1.tight_layout()
```

The external current must pass a threshold set by $V_\infty = V_{thresh}$ in order to spike.  This is
$$I_{ext} = \frac{V_{thresh} - E_m}{R} = 1.5 \mu A.$$
(Over a long enough time, machine precision would cause the neuron to spike at this current as my condition is $\ge$.)
As $I_{ext}$ increases, it increases $V_\infty$ and this increases the rate at which the voltage rises, thus increasing our spiking frequency.

### 1.3

```python
Is2 = scp.array(scp.arange(0, 11), dtype='float')
spike_freq = scp.zeros_like(Is2)
for idx, I_ext in enumerate(Is2):
    V, spike_times = V_LIF(V_reset, I_ext, t)
    if len(spike_times) <= 1:
        spike_freq[idx] = 0.
        #print 'Not enough spikes'
    else:
        spike_freq[idx] = 1.0 / (spike_times[1] - spike_times[0])
        #print idx, spike_freq[idx], spike_times[1] - spike_times[0]
fig2 = plt.figure()
ax2 = fig2.add_axes([0.1, 0.1, 0.8, 0.8])
ax2.plot(Is2, scp.zeros_like(Is2), 'k-')
ax2.plot(Is2, spike_freq)

ax2.set_ylim([-0.01, 0.5])
ax2.set_title(r'Spiking Frequency for LIF Neurons under varying $I_{ext}$')
ax2.set_xlabel(r'$I_{ext}$')
ax2.set_ylabel('Spiking Frequency (1/ms)')
```

### 1.4

```python
def V_LIF_ref(V_0, t_ref, I_ext, t):
    spike_times = []
    ref = scp.zeros(t_ref/dt)
    V = V_nospikes(V_0, I_ext, t)
    while any(V >= V_th):
        spk_idx = first_spike(V)
        spike_times.append(t[spk_idx])
        after_ref_idx = spk_idx + len(ref)
        if after_ref_idx > len(V)-1:
            V[spk_idx:] = scp.zeros_like(V[spk_idx:])
            return V, spike_times
        else:
            V[spk_idx:after_ref_idx] = ref
            V[after_ref_idx:] = V_nospikes(V_reset, I_ext, t[after_ref_idx:] - t[after_ref_idx])
    return V, spike_times
```

```python
Is2 = scp.array(scp.arange(0, 11), dtype='float')
t_ref_array = scp.arange(1, 6)
spike_freq = scp.zeros_like(Is2)
fig3 = plt.figure()
ax3 = fig3.add_axes([0.1, 0.1, 0.8, 0.8])
ax3.plot(Is2, scp.zeros_like(Is2), 'k-')
for t_ref in t_ref_array:
    for idx, I_ext in enumerate(Is2):
        V, spike_times = V_LIF_ref(V_reset, t_ref, I_ext, t)
        if len(spike_times) <= 1:
            spike_freq[idx] = 0.
            #print 'Not enough spikes'
        else:
            spike_freq[idx] = 1.0 / (spike_times[1] - spike_times[0])
            #print idx, spike_freq[idx], spike_times[1] - spike_times[0]
    ax3.plot(Is2, spike_freq, label=r'$t_{ref}=$'+str(t_ref))

ax3.set_ylim([-0.01, 0.35])
ax3.set_title(r'Spiking Frequency for LIF Neurons under varying $I_{ext}$ and $t_{ref}$')
ax3.set_xlabel(r'$I_{ext}$')
ax3.set_ylabel('Spiking Frequency (1/ms)')
ax3.legend(loc='upper left')
```

While without a refractory period the spiking frequency appears to increase linearly, once a refractory period is added the frequencies tend to saturate with high currents.  As increasing $I_{ext}$ takes $V_\infty$ up, the rate at which the voltage increases also climbs and the spikes become 'skinnier', increasing the frequency.  The refractory period sets a limit on how much it can increase.

### Bonus

$V_\infty$ is unbounded above and as our spike width ($w_{spike}$) (without a refractory period) approaches 0 as $V_\infty\to\infty$, we also have that the spike frequency is unbounded.  However, when we add a refractory period, the period of a spike becomes $t_{ref} + w_{spike}$ and as $I_{ext} \to \infty$ this approaches $t_{ref}$.  The maximum spiking frequency is thus $\frac{1}{t_{ref}}$.

### 1.5

```python
V_m = scp.linspace(V_reset, V_th, 100)
def deriv_V_t(V, I_ext):
    return (1 / C) * (I_ext - 1. / R * (V - E_m))

fig4 = plt.figure()
ax4 = fig4.add_axes([0.1, 0.1, 0.8, 0.8])

ax4.plot(V_m, scp.zeros_like(V_m), 'k')

for I in scp.arange(0, 11):
    ax4.plot(V_m, deriv_V_t(V_m, I), label=r'$I_{ext}=$'+str(I))

ax4.set_title('Phase-space Diagram for LIF Neuron')
ax4.set_xlabel('V (mV)')
ax4.set_ylabel(r'$\frac{dV}{dt}$ ($\frac{mV}{ms}$)')
```

Higher phase space trajectories correspond to progressively higher external currents.  For the two lowest currents ($I_{ext} = 0, 1$) there is a stable fixed-point where it crosses the $\frac{dV}{dt}=0$ line.  These correspond to stable voltages where the neuron will remain without spiking.  If the neuron is begun at $V_{reset} \le V(0) \le V_{th}$ it will seek out this stable point.

```python
g_GAP = 1                  # mS

def I_gap(V_self, V_other):
    return g_GAP * (V_self - V_other)

def I_L(V):
    return (1. / R) * (V - E_m)

def I_ext(t):
    if t <= 30:
        return scp.array([5, 3])
    elif t <= 70:
        return scp.array([3, 3])
    else:
        return scp.array([3, 0])

def deriv_V_t(V_12, t):
    I_ext_t = I_ext(t)
    dV_1 = (1. / C) * (I_ext_t[0] - I_L(V_12[0]) - I_gap(V_12[0], V_12[1]))
    dV_2 = (1. / C) * (I_ext_t[1] - I_L(V_12[1]) - I_gap(V_12[1], V_12[0]))
    return scp.array([dV_1, dV_2])
```

In order to get the integration of an event driven process, I think I might have to write an RK4 implementation (or at least that's the easiest way I think)

```python
def RK4_LIF(deriv, init, t_vec):
    V = scp.zeros((len(t_vec), 2))
    V[0,:] = init
    current_v = scp.zeros_like(init)
    next_v = scp.zeros_like(init)
    k = scp.zeros_like(init)
    for idx, time in enumerate(t_vec[:-1]):
        current_v = V[idx, :]
        h = t_vec[idx + 1] - time
        next_v = current_v + 1. / 6. * h * (k1(deriv, current_v, time) + 2 * k2(deriv, current_v, time, h) +
                                            2 * k3(deriv, current_v, time, h) + k4(deriv, current_v, time, h))
        next_v[next_v >= V_th] = V_reset
        V[idx + 1, :] = next_v
    return V
        

        
def k1(deriv, y, t):
    return deriv(y, t)

def k2(deriv, y, t, h):
    return deriv(y + 0.5 * h * k1(deriv, y, t), t + 0.5 * h)

def k3(deriv, y, t, h):
    return deriv(y + 0.5 * h * k2(deriv, y, t, h), t + 0.5 * h)

def k4(deriv, y, t, h):
    return deriv(y + h * k3(deriv, y, t, h), t + h)
```

For the first time period, the external currents are enough to cause either neuron to spike, however because neuron A spikes first the sudden voltage drop pulls down neuron B through the gap junction before it can spike.  This is because of the large size of $g_{GAP}$.  If the voltage difference between the two neurons is larger than $5mV$, the current through the gap junction is larger than either external current and will dominate.

In the second time period, both neurons have the same external current.  The gap junction causes them to sink up and they spike together. When $I_{ext2}$ is turned off, the current into neuron A is still enough to cause it to spike.  It also causes neuron B to follow it, raising the voltage of B and slowing the rise of neuron A.  As neuron B has no external current, it acts as extra 'inertia' slowing the spiking of neuron A.

### 2.2

We have
$$\frac{dV_A}{dt} = \frac{1}{C}(I_{extA} - \frac{1}{R}(V_A - E_m) - g_{GAP}(V_A - V_B))$$
and
$$\frac{dV_B}{dt} = \frac{1}{C}(I_{extB} - \frac{1}{R}(V_B - E_m) - g_{GAP}(V_B - V_A))$$

To derive the nullclines, we set these equal to zero, giving us linear equations in $V_A$ and $V_B$ for the points at which their respective derivatives are zero.  Setting the first equation equal to zero and solving for $V_B$,

$$0 = I_{extA} - \frac{1}{R}(V_A - E_m) - g_{GAP}(V_A - V_B)$$
$$V_B = (1 + \frac{1}{g_{GAP}R})V_A - (E_m + \frac{I_{extA}}{g})$$
Solving similarly for the second nullcline,
$$V_B = \frac{1}{1 + \frac{1}{g_{GAP}R}}V_A + \frac{(RI_{extB} + E_m)}{g_{GAP}R}$$

```python
def nullcline_V_A(V_A, I_extA):
    return (1. + (1. / (g_GAP * R))) * V_A - (R * I_extA + E_m) / (g_GAP * R)

def nullcline_V_B(V_A, I_extB):
    return (1. / (1. + 1. / (g_GAP * R))) * V_A + (R * I_extB + E_m) / (g_GAP * R)
```

```python
V_A = scp.linspace(0, 20, 100)

fig6, ax6 = subplots()
ax6.plot(V_A, nullcline_V_A(V_A, 3))
ax6.plot(V_A, nullcline_V_B(V_A, 5))
ax6.set_title('Nullclines of coupled LIF neurons')
ax6.set_xlabel(r'$V_A$')
ax6.set_ylabel(r'$V_B$')
```

Here we see that there are no stationary points of the coupled neurons. Given the currents, both neurons will spike, as we saw in the first time period of the above simulation.
