### Problem 1

We will begin with the reduced 2 variable Morris-Lecar model in which $w$ takes the place of $n$ in our potassium gating, we take the limit $\tau_m \to 0$ or perhaps $\tau_m \ll \tau_w$ so that our calcium (instead of sodium) gates are always at their equilibrium distribution.

```python
#Defining some parameters
C_m = 1                                     # uF/cm^2

E_Ca = 100                                  # mV
E_K = -70
E_L = -50

g_Ca = 1.1                                  # mS/cm^2
g_K = 2.0
g_L = 0.5
```

```python
# shifted to the left so it "leads" the dynamics and twice as steep in V
def m_inf(V): return (1 + scp.tanh((V + 1. ) / 15.)) / 2.
# a centered at V=0 and not as sharply increasing
def w_inf(V): return (1 + scp.tanh(V / 30.)) / 2.
# timescale is peaked at V=0 and decreases exponentially
def tau_w(V): return 5. / scp.cosh(V / 60)
def I_Ca(V):   return g_Ca * m_inf(V) * (V - E_Ca)
def I_K(V, w): return g_K * w * (V - E_K)
def I_L(V):    return g_L * (V - E_L)
```

```python
def deriv_V_t(V, w, I_ext):
    return (1. / C_m) * (-I_Ca(V) - I_K(V, w) - I_L(V) + I_ext)

def deriv_w_t(V, w):
    return (w_inf(V) - w) / tau_w(V)
```

This model is centered about $V=0$ and $w$ must run from 0 to 1 in our dynamics.

```python
V_lim = (-100, 100)
w_lim = (-0.1, 1)
V = scp.linspace(V_lim[0], V_lim[1])
w = scp.linspace(w_lim[0], w_lim[1])
V_grid, w_grid = scp.meshgrid(V, w)
```

```python
# Calculate function over grid
I = [5, 24.3, 35]
figs, axes = plt.subplots(len(I),1, figsize = (8, 12))
fixed_points = []
for I_ext, ax in zip(I, axes):

    def ML(VwIext, t):
        V, w = VwIext[0], VwIext[1]
        return deriv_V_t(V, w, I_ext), deriv_w_t(V, w)

    DVdt = deriv_V_t(V_grid, w_grid, I_ext)
    Dwdt = deriv_w_t(V_grid, w_grid)

    # Get Nullclines from function and crossings of Null_Clines
    nc1_V, nc1_w = getNullcline(DVdt, V, w)
    nc2_V, nc2_w = getNullcline(Dwdt, V, w)
    fp_V, fp_w = getCrossings(nc1_V, nc1_w, nc2_V, nc2_w)


    ax.plot(nc1_V, nc1_w, 'b-', label=r'$\frac{dV}{dt}=0$')
    ax.plot(nc2_V, nc2_w, 'g-', label=r'$\frac{dw}{dt}=0$')
    # Plot the fixed points
    
    t = np.linspace(0, 20)
    for fp_1, fp_2 in zip(fp_V, fp_w):
        fixed_points.append((fp_1, fp_2))
        ax.scatter(fp_1, fp_1)
        for init_num in range(5):
            initV = fp_1 + (2 * random.random()-1) * (200 * 0.01)
            initw = fp_2 + (2 * random.random() - 1) * (1.1 * 0.01)
            traj= odeint(ML, (initV, initw), t)
            ax.plot(traj[:,0], traj[:,1], 'r-')
                           
    ax.quiver(V_grid[::3], w_grid[::3], DVdt[::3], Dwdt[::3], angles='xy', scale_units='xy', scale=20 )
    ax.set_xlim([-100, 100])
    ax.set_ylim([-0.1, 1])
    if I_ext == 5:
        ax.legend()
    ax.set_title('Nullclines in Morris-Lecar Model for '+ repr(I_ext) + r' $uA/cm^2$ External Current')
    ax.set_ylabel('w')
    ax.set_xlabel('$V_m$ (mV)')

figs.tight_layout()
```

We can see that the first two fixed points are stable while the third is unstable and instead has a stable limit cycle.

```python
for (fp_1, fp_2), I_ext in zip(fixed_points, I):

    J = getJacobian(V_grid, w_grid, DVdt, Dwdt, fp_1, fp_2)
    [eig_val, eig_vec] = lin.eig(J)
    print 'Eigenvalues for I_ext = '+str(I_ext)+' are', eig_val
```

We can see that the first two are indeed stable (although the second is only marginally so and if I generate the points a little farther away, we do see trajectories escaping) and the third is unstable (although again, one of the directions is only marginally unstable.  This is the direction where the two nullclines run nearly parallel.

If we increased the external current to 200 $\mu A/cm^2$ we would see a single stable fixed point.  Physically (or maybe neurodynamically)  we can reason that at this voltage all of the $w$ and $m$ gates would be open and the system would revert to GHK dynamics with a large external current.  This is a linear system with negative feedback and a large potassium current would balance the internal current.
```python
C_m  =   1.0 # membrane capacitance, in uF/cm^2
g_Na =   120.0 # maximum conducances, in mS/cm^2
g_K  =   36.0
g_L  =   0.3
E_Na = 45.0 # Nernst reversal potentials, in mV
E_K  = -82.0
E_L  = -59.387
V_rest = -70.  # Resting potential

# Set up our transition coefficients
def alpha_m(V): return 0.1*(V + 45) / (1 - np.exp(-1.*(V + 45)/10))
def beta_m(V):  return 4 * np.exp(-1.*(V + 70) / 18)

def alpha_n(V): return 0.01*(V + 60) / (1 - np.exp(-1.*(V + 60)/10))
def beta_n(V):  return 0.125 * np.exp(-1.*(V + 70) / 80)

def deriv_n_t(V, n): return alpha_n(V) * (1 - n) - beta_n(V) * n
```

Now to set up the equilibrium conditions for each and making the approximation that $h = 0.8 - n$,

```python
def h(n):       return 0.8 - n
def m_infty(V): return alpha_m(V) / (alpha_m(V) + beta_m(V))
```

```python
def I_K(V, n):    return g_K * n**4 * (V - E_K)
def I_Na(V, n):   return g_K * m_infty(V)**4 * h(n) * (V - E_Na)
def I_L(V):       return g_L * (V - E_L)

def derivHH_V_t(V, n, I_ext):
    return (1. / C_m) * (-I_Na(V, n) - I_K(V, n) - I_L(V) + I_ext)

V_lim = (-200, 50)
n_lim = (0, 1)
V = scp.linspace(V_lim[0], V_lim[1], 100)
n = scp.linspace(n_lim[0], n_lim[1], 100)
V_grid, n_grid = scp.meshgrid(V, n)
```

We can now basically cut and paste the code above and obtain the same analysis

```python
I2 = [0, 5, 25, 125, 625]
figs2, axes2 = plt.subplots(len(I2),1, figsize = (8, 16))
fixed_points = []

for I_ext, ax in zip(I2, axes2):

    def HH(Vn, t):
        V, n = Vn[0], Vn[1]
        return derivHH_V_t(V, n, I_ext), deriv_n_t(V, n)

    DVdt = derivHH_V_t(V_grid, n_grid, I_ext)
    Dndt = deriv_n_t(V_grid, n_grid)

    # Get Nullclines from function and crossings of Null_Clines
    nc1_V, nc1_n = getNullcline(DVdt, V, n)
    nc2_V, nc2_n = getNullcline(Dndt, V, n)
    fp_V, fp_n = getCrossings(nc1_V, nc1_n, nc2_V, nc2_n)


    ax.plot(nc1_V, nc1_n, 'b-', label=r'$\frac{dV}{dt}=0$')
    ax.plot(nc2_V, nc2_n, 'g-', label=r'$\frac{dn}{dt}=0$')
    # Plot the fixed points
    
    t = np.linspace(0, 10)
    for fp_1, fp_2 in zip(fp_V, fp_n):
        fixed_points.append((fp_1, fp_2))
        ax.scatter(fp_1, fp_1)
        for init_num in range(5):
            initV = fp_1 + (2 * random.random() - 1) * (250 * 0.01)
            initn = fp_2 + (2 * random.random() - 1) * (1 * 0.01)
            traj= odeint(HH, (initV, initn), t)
            ax.plot(traj[:,0], traj[:,1], 'r-')
                           
    ax.quiver(V_grid[::8], n_grid[::8], DVdt[::8], Dndt[::8], angles='xy', scale_units='xy', scale=100 )
    ax.set_xlim([-200, 50])
    ax.set_ylim([0, 1])
    if I_ext == 5:
        ax.legend()
    ax.set_title('Nullclines in Hodgkin-Huxley Model for '+ repr(I_ext) + r' $uA/cm^2$ External Current')
    ax.set_ylabel('n')
    ax.set_xlabel('$V_m$ (mV)')

figs2.tight_layout()
```

```python
for (fp_1, fp_2), I_ext in zip(fixed_points, I2):

    J = getJacobian(V_grid, n_grid, DVdt, Dndt, fp_1, fp_2)
    [eig_val, eig_vec] = lin.eig(J)
    print 'Eigenvalues for I_ext = '+str(I_ext)+' are', eig_val
```
