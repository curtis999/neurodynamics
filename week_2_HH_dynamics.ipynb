## Overview

We now consider the effect of gated channels on ion flow. Each channel contains a series of gates, 4 in the case of potassium, and 3+1 for sodium.  All of these gates must be open for the channel to conduct.  We imagine the gates to be controlled by a markov process with transition coefficients that depend on the membrane potential.  These transitions reach an equilibrium for a set voltage with timescales of $.1-1ms$ which is fast compared to most membrane dynamics.  Hodgkin and Huxley realized that by having gates that "turned on" (and off) at different voltages, we can generate action potentials.  The process is as follows:

1. At rest we have mostly potassium flow out of the cell countering a leak current.
2. As the potential rises, the sodium current turns on first, quickly pulling the membrane potential up (these are the 3 gates in the 3+1)
3. As the potential rises the other gate in sodium whose equilibrium value is a decreasig function of voltage begins to turn off, abruptly cutting off the flow of sodium and allow potassium to restore the cell to its resting potential

It is also worth noting that this occurs only at the thin border of the membrane and the number of ions that flow across is a small fraction of the total in the bulk.  The cell is "charged" enough to give many action potentials without the operation of its ion pumps and the dynamics occur only along the membrane.

### Ion Gated Channels

Ion exchange through a membrane is mediated by channels whose flow is controlled by a set of gates.  These are depicted below

HH found that their data was best fit by including 4 gating variables in the potassium channels and 3 fast and 1 slow gating variables in the sodium channels.  The meaning of fast and slow will become clear later.  The gates are essentially like little doors that, all of which must be open in a given channel in order for current to flow.  As the gate's transition rates vary with the membrane potential, they can give positive/negative feedback to changes and give rise to interesting dynamics.

### Channel Gate Dynamics

We can consider a single gate's state to be a stochastic variable (which can be either open or closed, or 1 or 0).  Across a whole neuron, we consider the fraction of gates open and closed and by the CLT this smooths our dynamics.  Defining $n$ as the fraction of open gates, $\alpha_n$ as the opening rate and $\beta_n$ as the closing rate, we then have, in the continuum limit,
$$\frac{dn}{dt} = \alpha_n(V_m) (1-n) - \beta_n(V_m)n.$$
We rearrange this to clarify the meaning of some terms:
$$\tau_n\frac{dn}{dt} = n_\infty(V_m) - n$$
which gives a timescale $\tau_n = \frac{1}{\alpha_n + \beta_n}$ and equilibrium value $n_\infty = \frac{\alpha_n}{\alpha_n + \beta_n}$. The solution of this at a particular voltage is simply
$$(n_0 - n_\infty) e^{-t/\tau} + n_\infty$$
so the system exponentially approches the equilibrium value with timescale $\tau$.

The dynamics suggested by this are fairly boring unless $\alpha$ and $\beta$ vary with $V_m$ (why else would you have gone through the work to evolve gates). To get at the form of this dependence, we imagine the gate as a two state system with energies $\epsilon_o$ and $\epsilon_c$ associated with being open or closed respectively. At equilibrium, the two states will occur in proportion to their boltzmann factors: $\frac{n_o}{n_c} = e^{-(\epsilon_o - \epsilon_c)/kT}$.  Similarly, at equilibrium, transitions between the states must preserve this distribution so $\alpha n_c = \beta n_o$ and so we have
$$\frac{\alpha}{\beta} = e^{-(\epsilon_o - \epsilon_c)/kT}$$.

### Problem 1

1. We want to solve $\tau\frac{dn}{dt} + n = n_\infty = \frac{\alpha}{\alpha + \beta}$. This is a 1st order, inhomogeneous differential equation.  The solution to the homogeneous equation is $n(t) = Ae^{-t/\tau}$ and the particular solution is $n_p = \frac{\alpha}{\alpha + \beta}$.  Assuming an initial condition $n_0$ we have a general solution,
$$n(t) = (n_0 - n_\infty) e^{-t/\tau} + n_\infty$$
The system exponentially approaches its equilibrium value with time constant $\tau = \frac{1}{\alpha+\beta}$

For the given form of $V_m = 30*H(t-t_{input})$ and $\alpha(V_m) = \alpha_0 e^{\lambda V_m} \quad \beta(V_m) = \beta_0 e^{-\mu V_m}$, the solution is:

from t = 0 to t = 8,
$$n(t) = \frac{\alpha_0}{\alpha_0 + \beta_0}( 1 -  e^{-t/\tau})$$.
And calling $n(8) = n_8$,
from t = 8 to t = $\infty$,
$$n(t) = (n_8 - \frac{\alpha_0 e^{\lambda V_m}}{\alpha_0 e^{\lambda V_m} + \beta_0 e^{-\mu V_m}}) e^{-(t-8)/\tau} + \frac{\alpha_0 e^{\lambda V_m}}{\alpha_0 e^{\lambda V_m} + \beta_0 e^{-\mu V_m}}$$

```python
#setting up parameters
alpha_0 = 0.7 # Hz
beta_0 = 0.9

q = 0.01 # 1/mV
t_input = 8 # sec


def alpha(V):
    return alpha_0 * np.exp(q * V)

def beta(V):
    return beta_0 * np.exp(-q * V)

def V_membrane(t):
    return 30 * (t > t_input)

def n_analytic(t, n_0):
    V = V_membrane(t)
    n_inf = alpha(V) / (alpha(V) + beta(V))
    tau = 1 / (alpha(V) + beta(V))
    if t <= 8:
        return (n_0 - n_inf) * np.exp( -t / tau ) + n_inf
    else:
        n_8 = n_analytic(8, n_0)
        return (n_8 - n_inf) * np.exp( -(t - 8) / tau ) + n_inf
    
def n_analytic_from_0(t):
    return n_analytic(t, 0)


def deriv_n_t(n, t):
    V = V_membrane(t)
    return alpha(V)* (1 - n) - beta(V) * n

t = np.arange(0, 16, 10**-2)

n_ode = scp.integrate.odeint(deriv_n_t, 0, t)
n = map(n_analytic_from_0, t)
```

```python
def fraction(state):
    return sum(state)/len(state)

# an attempt to refactor this into something nicer and faster
def n_stochastic(initial, t):
    # initialize our variables
    dt = t[1] - t[0]
    n = np.zeros_like(t)                            # Need an extra step to include the endpoint
    state = np.zeros_like(initial)                  # holds the current state
    n[0] = fraction(initial)                        # store the inital value in the zero slot
    
    # step through the timepoints
    for (step,), time in np.ndenumerate(t[:-1]):
        
        rand = np.random.random_sample(initial.shape)
        # Step through the gates
        for (idx,), val in np.ndenumerate(state):
            # if the gate is closed, see if it opens
            if val == 0 and alpha(V_membrane(time))*dt > rand[idx]:
                state[idx] = 1
            # and if it's open see if it closed
            elif val == 1 and beta(V_membrane(time))*dt > rand[idx]:
                state[idx] = 0
        n[step+1] = fraction(state)
    return n

ns1 = n_stochastic(np.zeros(1), t)
ns400 = n_stochastic(np.zeros(400), t)
ns1000 = n_stochastic(np.zeros(1000), t)
```

```python
fig1 = plt.figure(figsize = (8, 6))
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(t, n, 'k-', label='Analytic')
axes1.plot(t, n_ode, 'g--', label='Numerical')
axes1.plot(t, ns1, 'r-', label='Stoch n=1')
axes1.plot(t, ns400, 'y-', label='Stoch n=400')
axes1.plot(t, ns1000, 'b-', label='Stoch n=1000')
axes1.legend()
axes1.set_xlabel('Time (s)')
axes1.set_ylabel('Fraction of gates open')
axes1.set_title('Fraction of open gates for various simulation schemes')
```

### Problem 2

```python
C_m  =   1.0 # membrane capacitance, in uF/cm^2
g_Na =   120.0 # maximum conducances, in mS/cm^2
g_K  =   36.0
g_L  =   0.3
E_Na = 45.0 # Nernst reversal potentials, in mV
E_K  = -82.0
E_L  = -59.387
V_rest = -70.  # Resting potential

# Set up our transition coefficients
def alpha_m(V): return 0.1*(V + 45) / (1 - np.exp(-1.*(V + 45)/10))
def beta_m(V):  return 4 * np.exp(-1.*(V + 70) / 18)

def alpha_h(V): return 0.07 * np.exp(-1.*(V + 70) / 20)
def beta_h(V):  return 1. / (1 + np.exp(-1.*(V + 40)/10))

def alpha_n(V): return 0.01*(V + 60) / (1 - np.exp(-1.*(V + 60)/10))
def beta_n(V):  return 0.125 * np.exp(-1.*(V + 70) / 80)

V = np.linspace(-90, 70)

fig2,axes2 = plt.subplots(1,3, figsize=(12,4))

gate_type = ['n', 'm', 'h']

for ax,letter in zip(axes2, gate_type):
    ax.set_xlabel ('Potential (mV)')
    ax.set_title(letter)
    ax.set_xlim([-90, 70])

axes2[0].plot(V, alpha_n(V), 'b-', label='$\alpha_n$')
axes2[0].plot(V, beta_n(V), 'b--', label='$\beta_n$')
axes2[1].plot(V, alpha_m(V), 'r-', label='$\alpha_m$')
axes2[1].plot(V, beta_m(V), 'r--', label='$\beta_m$')
axes2[2].plot(V, alpha_h(V), 'g-', label='$\alpha_h$')
axes2[2].plot(V, beta_h(V), 'g--', label='$\beta_h$')

fig2.tight_layout()
```

For some reason this plot wouldn't let me add legends.  As such, in each plot, $\alpha$ is a solid line and $\beta$ is dashed.

```python
def I_leak(V): return g_L * (V - E_L)

#I_ext steps up by 10 uA every 100 ms
def I_ext(t):
    return 10 * np.floor(t/100.)

def deriv_V_t_leak_ext(V, t):
    return (1. / C_m) * (-I_leak(V) + I_ext(t))

t = np.linspace(0, 500, num=500)
V_ext_leak = scp.integrate.odeint(deriv_V_t_leak_ext, V_rest, t)

fig3 = plt.figure()
ax3 = fig3.add_axes([0.1, 0.1, 0.8, 0.8])
ax3.plot(t, V_ext_leak, 'k-')
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Membrane Potential')
ax3.set_title('Membrane under increasing external current with only $I_L$')
```

So each $uA$ increases the membrane potential by about 3.5 mV.

```python
def I_Na(V, m, h):  return g_Na * m**3*h * (V - E_Na)
def I_K(V, n):   return g_K  * n**4   * (V - E_K)

def deriv_n_t(V, n): return alpha_n(V) * (1 - n) - beta_n(V) * n
def deriv_m_t(V, m): return alpha_m(V) * (1 - m) - beta_m(V) * m
def deriv_h_t(V, h): return alpha_h(V) * (1 - h) - beta_h(V) * h

def HH(Vnmh, t):
    V, n, m, h = Vnmh[0], Vnmh[1], Vnmh[2], Vnmh[3]
    
    dV = (1. / C_m) * (-I_Na(V, m, h) - I_K(V, n) - I_leak(V) + I_ext(t))
    dn = deriv_n_t(V, n)
    dm = deriv_m_t(V, m)
    dh = deriv_h_t(V, h)
    
    return dV, dn, dm, dh

# Letting the HH model come to a steady state without an external current
# has yielded the following gating variable values
n0 = 0.31773
m0 = 0.05296
h0 = 0.59599

t = linspace(0, 500, 1000)
V_membrane = scp.integrate.odeint(HH, (V_rest, n0, m0, h0), t)

fig4 = plt.figure()
ax4 = fig4.add_axes([0.1, 0.1, 0.8, 0.8])
ax4.plot(t, V_membrane[:,0])
ax4.set_ylim([-100, 10])
ax4.set_xlabel('Time (ms)')
ax4.set_ylabel('V_m (mV)')
ax4.set_title('V_membrane subject to an increasing external current')
```

We see that the neuron begins spiking wildly after we increase the external current past $10\mu A$ and the frequency of spike grows with the external current, however the amplitude of the spikes begins to decay.  To see what happens when we increase it past $160 \mu A$ we solve the system for a longer time.

```python
t_long = np.linspace(0, 1800, 1800)

V_long = scp.integrate.odeint(HH, (V_rest, n0, m0, h0), t_long)
```

```python
fig5 = plt.figure()
ax5 = fig5.add_axes([0.1, 0.1, 0.8, 0.8])
ax5.plot(t_long, V_long[:,0])
ax5.set_ylim([-100, 10])
ax5.set_xlabel('Time (ms)')
ax5.set_ylabel('V_m (mV)')
ax5.set_title('A HH neuron under increasing external current')
```

The graph is not so transparent over such a long time scale, but we do see the voltage entering another fixed point around -50mV.
```python
fig6 = plt.figure()
ax6 = fig6.add_axes([0.1, 0.1, 0.8, 0.8])
ax6.plot(t[380:460], V_membrane[380:460,0])
ax6.set_ylim([-100, 10])
ax6.set_xlabel('Time (ms)')
ax6.set_ylabel('V_m (mV)')
ax6.set_title('V_membrane subject to an increasing external current')
```

It's difficult to tell precisely, but it certainly seems to be on the order of 5-10 mV. The graph seems to turn sharply when the voltage reaches -60mV.

Looking at the transition coefficients, the order of magnitude imbalance between potassium and sodium transition coefficients is responsible for the runaway potential in the spike. By shifting both coefficients for both ions to the right, we should be able to change the point at which this runaway occurs. ### Problem 3

#### 3.1

We have $n_4 = n^4$ if they obey the same differential equation.  We have
$$\frac{dn_4}{dt} = \alpha n_3 - 4 \beta n_4$$
and
$$\frac{dn}{dt} = \alpha (1-n) - \beta n.$$
Then
$$\frac{d}{dt}n^4 = 4 n^3 \frac{dn}{dt} = 4\alpha n^3 (1-n) - 4 \beta n^4$$
Identifying $n_3 = 4 n^3 (1-n)$ we have
$$\frac{dn^4}{dt} = \alpha n_3 - 4\beta n_4$$

#### 3.2

Same drill.  We have
$$\frac{ds_{31}}{dt} = \alpha_m s_{21} - 3\beta_m s_{31} + \alpha_h s_{30} - \beta_h s_{31}$$
$$\frac{dm}{dt} = \alpha_m (1-m) - \beta_m m$$
and
$$\frac{dh}{dt} = \alpha_h (1-h) - \beta_h h.$$
Checking,
$$\frac{d}{dt} m^3h = 3m^2h\frac{dm}{dt} + m^3\frac{dh}{dt} = \alpha_m 3m^2(1-m)h - \beta_m 3m^3h + \alpha_h m^3(1-h) - \beta_h m^3h$$
Making similar combinatoric identities to the previous problem, this simplifies to
$$\frac{d}{dt} m^3h =\alpha_m s_{21} - 3\beta_m s_{31}  + \alpha_h s_{30} - \beta_h s_{31}$$

#### 3.3

To do this, we will evolve the gates and Voltage on two different time scales.  The gate time scale should be divided by $\tau.$ The fastest time scale in our voltage range seems to be approximately 1/4ms, so the stochastic part will be evolved on a 1/10 ms timescale. Between each stochastic step, we will evolve $V$ as an ode for 10 steps.

```python
def I_ext_stoch(t):
    return 2 * np.floor(t/50.)

def dV_dt_stoch(V, t, n, m, h,):
    return (1. / C_m) * (-I_Na(V, m, h) - I_K(V, n) - I_leak(V) + I_ext_stoch(t))


t_stoch = np.arange(0, 400, 0.1)
dt_stoch = t_stoch[1] - t_stoch[0]

n0 = 0.31773
m0 = 0.05296
h0 = 0.59599
#number of each type of channel.  This will make sure the relative size of each moment is correct
num_channels = 500
num_n_gates = 4 * num_channels
num_m_gates = 3 * num_channels
num_h_gates = 1 * num_channels

n_open = np.round(n0 * num_n_gates)
n_closed = num_n_gates - n_open
m_open = np.round(m0 * num_m_gates)
m_closed = num_m_gates - m_open
h_open = np.round(h0 * num_h_gates)
h_closed = num_h_gates - h_open


n_gates = np.concatenate((np.ones(n_open), np.zeros(n_closed)))
m_gates = np.concatenate((np.ones(m_open), np.zeros(m_closed)))
h_gates = np.concatenate((np.ones(h_open), np.zeros(h_closed)))

n_stoch = np.zeros_like(t_stoch)
m_stoch = np.zeros_like(t_stoch)
h_stoch = np.zeros_like(t_stoch)
V_stoch = np.zeros_like(t_stoch)

n_stoch[0] = n0
m_stoch[0] = m0
h_stoch[0] = h0
V_stoch[0] = V_rest
```

```python
# step through the timepoints
for (step,), time in np.ndenumerate(t_stoch[:-1]):
    
    n_rand = np.random.random_sample(n_gates.shape)
    # Step through the gates
    for (nidx,), val in np.ndenumerate(n_gates):
        # if the gate is closed, see if it opens
        if val == 0 and alpha_n(V_stoch[step])*dt_stoch > n_rand[nidx]:
            n_gates[nidx] = 1
        # and if it's open see if it closed
        elif val == 1 and beta_n(V_stoch[step])*dt_stoch > n_rand[nidx]:
            n_gates[nidx] = 0
    n_stoch[step+1] = fraction(n_gates)
    
    m_rand = np.random.random_sample(m_gates.shape)
    # Step through the gates
    for (midx,), val in np.ndenumerate(m_gates):
        # if the gate is closed, see if it opens
        if val == 0 and alpha_m(V_stoch[step])*dt_stoch > m_rand[midx]:
            m_gates[midx] = 1
        # and if it's open see if it closed
        elif val == 1 and beta_m(V_stoch[step])*dt_stoch > m_rand[midx]:
            m_gates[midx] = 0
    m_stoch[step+1] = fraction(m_gates)
    
    h_rand = np.random.random_sample(h_gates.shape)
    # Step through the gates
    for (hidx,), val in np.ndenumerate(h_gates):
        # if the gate is closed, see if it opens
        if val == 0 and alpha_h(V_stoch[step])*dt_stoch > h_rand[hidx]:
            h_gates[hidx] = 1
        # and if it's open see if it closed
        elif val == 1 and beta_h(V_stoch[step])*dt_stoch > h_rand[hidx]:
            h_gates[hidx] = 0
    h_stoch[step+1] = fraction(h_gates)
    
    def dV_dt(V, t):
        return dV_dt_stoch(V, t, n_stoch[step], m_stoch[step], h_stoch[step])
    
    t_v = np.linspace(time, time + dt_stoch, 10)                       
    V_short = scp.integrate.odeint(dV_dt, V_stoch[step], t_v)
    
    V_stoch[step+1] = V_short[-1]
```
