### Problem 1

We will begin with the reduced 2 variable Morris-Lecar model in which $w$ takes the place of $n$ in our potassium gating, we take the limit $\tau_m \to 0$ or perhaps $\tau_m \ll \tau_w$ so that our calcium (instead of sodium) gates are always at their equilibrium distribution.

```python
#Defining some parameters
C_m = 1                                     # uF/cm^2

E_Ca = 100                                  # mV
E_K = -70
E_L = -50

g_Ca = 1.1                                  # mS/cm^2
g_K = 2.0
g_L = 0.5
```

```python
# shifted to the left so it "leads" the dynamics and twice as steep in V
def m_inf(V): return (1 + scp.tanh((V + 1. ) / 15.)) / 2.
# a centered at V=0 and not as sharply increasing
def w_inf(V): return (1 + scp.tanh(V / 30.)) / 2.
# timescale is peaked at V=0 and decreases exponentially
def tau_w(V): return 5. / scp.cosh(V / 60)
def I_Ca(V):   return g_Ca * m_inf(V) * (V - E_Ca)
def I_K(V, w): return g_K * w * (V - E_K)
def I_L(V):    return g_L * (V - E_L)
```

```python
def deriv_V_t(V, w, I_ext):
    return (1. / C_m) * (-I_Ca(V) - I_K(V, w) - I_L(V) + I_ext)

def deriv_w_t(V, w):
    return (w_inf(V) - w) / tau_w(V)
    
def get_init(fpx, fpy, N, xscale, yscale, length=10**-2):
    """
    Generates N initial conditions radially around the fixed point in V-w space
    """
    xlen = xscale * length
    ylen = yscale * length
    
    phasors = [np.exp(complex(0,(2 * np.pi * n / N))) for n in range(N)]
    points = np.zeros_like(phasors)
    for idx, phasor in enumerate(phasors):
        points[idx] = (fpx + xlen * phasor.real, fpy + ylen * phasor.imag)
    return points
```

This model is centered about $V=0$ and $w$ must run from 0 to 1 in our dynamics.

```python
V_lim = (-100, 100)
w_lim = (-0.1, 1)
V = scp.linspace(V_lim[0], V_lim[1])
w = scp.linspace(w_lim[0], w_lim[1])
V_grid, w_grid = scp.meshgrid(V, w)
```

```python
# Calculate function over grid
I_ext = 35

def ML(VwIext, t):
    V, w = VwIext[0], VwIext[1]
    return deriv_V_t(V, w, I_ext), deriv_w_t(V, w)

DVdt = deriv_V_t(V_grid, w_grid, I_ext)
Dwdt = deriv_w_t(V_grid, w_grid)

# Get Nullclines from function and crossings of Null_Clines
nc1_V, nc1_w = getNullcline(DVdt, V, w)
nc2_V, nc2_w = getNullcline(Dwdt, V, w)
fp_V, fp_w = getCrossings(nc1_V, nc1_w, nc2_V, nc2_w)

```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(nc1_V, nc1_w, 'b-', label=r'$\frac{dV}{dt}=0$')
ax.plot(nc2_V, nc2_w, 'g-', label=r'$\frac{dw}{dt}=0$')
# Plot the fixed points

t = np.linspace(0, 20)
for fp_1, fp_2 in zip(fp_V, fp_w):
    ax.scatter(fp_1, fp_1)
    for init_num in range(5):
        initV = fp_1 + (2 * random.random()-1) * 4
        initw = fp_2 + (2 * random.random() - 1) * 0.04
        traj= odeint(ML, (initV, initw), t)
        ax.plot(traj[:,0], traj[:,1], 'r-')
                           
ax.quiver(V_grid[::3], w_grid[::3], DVdt[::3], Dwdt[::3], angles='xy', scale_units='xy', scale=20 )
ax.set_xlim([-100, 100])
ax.set_ylim([-0.1, 1])
ax.legend()
ax.set_title('Nullclines in Morris-Lecar Model for '+ repr(I_ext) + r' $uA/cm^2$ External Current')
```
