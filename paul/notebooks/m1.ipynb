***Nearest-neighbor interactions***

```python
structure = 'rect_periodic'
M = 4  # number of rows
N = 4  # number of columns

# uniform connection strength
Jm = 1.0

J = np.zeros(shape=(M*N,M*N))
for ns in range(M*N):
    # get index numbers of neighbors
    iL = ns - 1
    iR = ns + 1
    iT = ns - N
    iB = ns + N
    
    # now modify edge oscillators to enforce periodic boundary conditions
    if ns%N == 0:  # left-hand edge
        iL = ns + N - 1
    if ns%N == (N-1):  # right-hand edge
        iR = ns - N + 1
    if ns < N:  # top edge
        iT = ns + N*(M-1)
    if ns > (N*M - 1 - N):  # bottom edge
        iB = (ns%N)
    
    # fill in connection strenghts
    J[ns,iL] = Jm
    J[ns,iR] = Jm
    J[ns,iT] = Jm
    J[ns,iB] = Jm

# initialize the state
x = np.array(np.random.randn(M*N))
```

### Run the simulation, collect time series data

```python
# equations of motion
def dxdt(x, t, gJ, J):
    return -x + np.einsum('ij,j', J, np.tanh(gJ*x))
```

```python
tstart = time.time()

t0 = 0.0
tf = 6000.0
dt = 0.1
Nt = (tf+dt - t0)/dt
times = np.arange(t0, tf+dt, dt)

sol = odeint(dxdt, x, times, (gJ,J))
x = sol[-1]

print time.time()-tstart, ' s'
```

### Calculating Lyapunov exponents

```python
import calc_LE
```

Here I take the model, and also calculate the matrix of derivatives
$$ \frac{\partial f_i(t+t_0)}{\partial x_j(t_0)} $$
which are all passed to the function calc_LE.local_LE.  This calculates a time series of the largest Lyapunov exponent; this should asymptote to the global value.  The point here is to test if the model, given $g$ and $J$, is chaotic.

```python
# calculate Df_ij
def Df_m1(x, pars):
    gJ = pars[0]  # control parameter
    J = pars[1]  # matrix of connections
    return -np.eye(N, dtype='float') + gJ*np.einsum('ik,jk', J, np.resize(1.0 - np.tanh(gJ*x)**2.0, (N,N)))
```

```python
tnow = time.time()
boom = Df_m1(x, (gJ,J))
print time.time()-tnow, ' s'
```

```python
# generate new initial condition, and integrate out the transient, if not already done
x0_a = np.random.randn(N)

t0 = 0.0
tf = 0.5
dt = 0.1
times = np.arange(t0,tf+dt,dt)

pars = (gJ,J)

x0_a = odeint(dxdt, x0_a, times, pars)[-1]
x0_b = x0_a + (0.001*np.random.randn(N))
```

```python
# else use current state as initial condition
x0_a = x
x0_b = x0_a + (0.0001*np.random.randn(N))
```

### Calculate correlation times, lengths

- Take the time series generated in the top section of code, and use it to calculate both
$$
\begin{align}
\Delta(t) &\equiv \left\langle h_i(t_0+t) h_i(t_0) \right\rangle \\
\Gamma(n) &\equiv \left\langle h_{i+n} h_i \right\rangle
\end{align}
$$
where, for now, the averages are taken over the system (eventually these should become ensemble averages).  The second quantity should probably be time-averaged; it will be a function of the "distance" across the network, something like $\sqrt{n_x^2 + n_y^2}$.


- There is, however, no notion of "distance" in the randomly connected (nonlocal) network.  To compare to the other cases, I think I will impose a grid structure on the network, solely based on the index of the individual component.  I think the correlation lengths should be roughly constant as a function of $n$ in the nonlocal case.  We'll see.

```python
# find autocorrelation as a function of time
tstart = time.time()
corr_vec = np.correlate(sol[:,0], sol[:,0], mode='full')# * np.append(-np.arange(-Nt, 0, 1), np.arange(2, Nt+1, 1))
#corr_vec /= np.sqrt(np.einsum('i,i', corr_vec, corr_vec))
#plt.plot(times, corr_vec[(len(corr_vec)-1)/2:])
plt.plot(corr_vec)
plt.show()
print time.time()-tstart, ' s'
```
